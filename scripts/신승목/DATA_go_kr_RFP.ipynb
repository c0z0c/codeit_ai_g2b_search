```python
import re
import requests
from typing import Optional

def get_data_go_kr(
    api_url: str,
    timeout: int = 10,
    service_key: Optional[str] = None
) -> dict:
    """
    ê³µê³µë°ì´í„°í¬í„¸ API ê³µí†µ í˜¸ì¶œ í•¨ìˆ˜.
    
    Args:
        api_url: API ì—”ë“œí¬ì¸íŠ¸ URL (serviceKey íŒŒë¼ë¯¸í„° ì œì™¸).
        timeout: íƒ€ì„ì•„ì›ƒ (ì´ˆ).
        service_key: ì„œë¹„ìŠ¤ í‚¤. Noneì´ë©´ í™˜ê²½ë³€ìˆ˜ì—ì„œ ë¡œë“œ.
    
    Returns:
        API ì‘ë‹µ JSON ë”•ì…”ë„ˆë¦¬.
    
    Raises:
        ValueError: í™˜ê²½ë³€ìˆ˜ ë¯¸ì„¤ì • ë˜ëŠ” API ì˜¤ë¥˜ ì‘ë‹µ.
        requests.HTTPError: HTTP ì˜¤ë¥˜.
    """
    if service_key is None:
        service_key = os.getenv('DATA_GO_KR_SERVICE_KEY')
    
    if not service_key:
        logger.error("DATA_GO_KR_SERVICE_KEY í™˜ê²½ë³€ìˆ˜ ë¯¸ì„¤ì •")
        raise ValueError("DATA_GO_KR_SERVICE_KEY í™˜ê²½ë³€ìˆ˜ê°€ ì„¤ì •ë˜ì§€ ì•Šì•˜ìŠµë‹ˆë‹¤")
    
    full_url = f"{api_url}&serviceKey={service_key}"
    logger.debug(f"API ìš”ì²­: {api_url[:100]}...")
    
    response = requests.get(full_url, timeout=timeout)
    response.raise_for_status()
    data = response.json()
    
    # ì‘ë‹µ êµ¬ì¡° ê²€ì¦
    if 'response' not in data:
        logger.error("ì‘ë‹µ êµ¬ì¡° ì˜¤ë¥˜: 'response' í‚¤ ì—†ìŒ")
        raise ValueError("API ì‘ë‹µ êµ¬ì¡° ì˜¤ë¥˜: 'response' í‚¤ê°€ ì—†ìŠµë‹ˆë‹¤")
    
    result_code = data.get('response', {}).get('header', {}).get('resultCode')
    result_msg = data.get('response', {}).get('header', {}).get('resultMsg')
    
    if result_code != '00':
        logger.error(f"API ì˜¤ë¥˜: code={result_code}, msg={result_msg}")
        raise ValueError(f"API ì˜¤ë¥˜ ì‘ë‹µ: [{result_code}] {result_msg}")
    
    return data


from datetime import datetime, timedelta
from typing import List, Dict

# https://apis.data.go.kr/1230000/ad/BidPublicInfoService/getBidPblancListPPIFnlRfpIssAtchFileInfo?serviceKey=8fea481d8c22a53e6baa58ac84350a16577b68d4ba8ee4b2c5ce3057db4bbbfa&pageNo=1&numOfRows=10&inqryDiv=1&type=json&bidNtceNo=R25BK01156560&inqryBgnDt=202506010000&inqryEndDt=202507010000
def get_bid_file_info(
    start_date: str, 
    end_date: str, 
    bid_ntce_no: str = '', 
    page_no: int = 1, 
    num_of_rows: int = 10, 
    inqry_div: int = 1
) -> List[Dict]:
    """
    ì…ì°° ê³µê³  ì²¨ë¶€ íŒŒì¼ ì •ë³´ë¥¼ ì¡°íšŒí•˜ë©°, 1ê°œì›” ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ì–´ í˜¸ì¶œ.

    Args:
        start_date: ì¡°íšŒ ì‹œì‘ ë‚ ì§œ (YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM).
        end_date: ì¡°íšŒ ì¢…ë£Œ ë‚ ì§œ (YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM).
        bid_ntce_no: ì…ì°° ê³µê³  ë²ˆí˜¸.
        page_no: í˜ì´ì§€ ë²ˆí˜¸.
        num_of_rows: í˜ì´ì§€ë‹¹ ê²°ê³¼ ìˆ˜.
        inqry_div: ì¡°íšŒ êµ¬ë¶„.

    Returns:
        List[Dict]: ë³‘í•©ëœ API ì‘ë‹µ ë°ì´í„° ë¦¬ìŠ¤íŠ¸.
        
    Raises:
        ValueError: ë‚ ì§œ í˜•ì‹ ì˜¤ë¥˜.
    """
    # ë‚ ì§œ í˜•ì‹ ê²€ì¦
    date_pattern = re.compile(r'^\d{8}(\d{4})?$')
    if not date_pattern.match(start_date):
        logger.error(f"start_date í˜•ì‹ ì˜¤ë¥˜: {start_date}")
        raise ValueError(f"start_dateëŠ” YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM í˜•ì‹ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    if not date_pattern.match(end_date):
        logger.error(f"end_date í˜•ì‹ ì˜¤ë¥˜: {end_date}")
        raise ValueError(f"end_dateëŠ” YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM í˜•ì‹ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    
    # YYYYMMDD â†’ YYYYMMDDHHMM ë³€í™˜
    if len(start_date) == 8:
        start_date += '0000'
    if len(end_date) == 8:
        end_date += '0000'
    
    # ë‚ ì§œ ë³€í™˜
    start_dt = datetime.strptime(start_date, "%Y%m%d%H%M")
    end_dt = datetime.strptime(end_date, "%Y%m%d%H%M")
    if start_dt > end_dt:
        raise ValueError("start_dateëŠ” end_dateë³´ë‹¤ ì´ì „ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    
    # ê²°ê³¼ ë³‘í•©
    merged_results = []
    current_start = start_dt

    while current_start < end_dt:
        # 1ê°œì›” ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ê¸°
        current_end = (current_start + timedelta(days=31)).replace(day=1) - timedelta(seconds=1)
        if current_end > end_dt:
            current_end = end_dt
        
        # ë‚ ì§œë¥¼ ë¬¸ìì—´ë¡œ ë³€í™˜
        current_start_str = current_start.strftime("%Y%m%d%H%M")
        current_end_str = current_end.strftime("%Y%m%d%H%M")
        
        logger.debug(f"API í˜¸ì¶œ: {current_start_str} ~ {current_end_str}")
        
        # API í˜¸ì¶œ
        try:
            data = get_data_go_kr(
                api_url=(
                    f"https://apis.data.go.kr/1230000/ad/BidPublicInfoService/"
                    # f"getBidPblancListPPIFnlRfpIssAtchFileInfo"
                    f"get_bid_file_info?"
                    f"?pageNo={page_no}"
                    f"&numOfRows={num_of_rows}"
                    f"&inqryDiv={inqry_div}"
                    f"&type=json"
                    f"&bidNtceNo={bid_ntce_no}"
                    f"&inqryBgnDt={current_start_str}"
                    f"&inqryEndDt={current_end_str}"
                ),
                timeout=10
            )
            items = data.get('response', {}).get('body', {}).get('items', [])
            if items:
                merged_results.extend(items)
                logger.info(f"{current_start_str} ~ {current_end_str}: {len(items)}ê±´ ì¡°íšŒ")
            else:
                logger.info(f"{current_start_str} ~ {current_end_str}: ì¡°íšŒëœ ë¬¸ì„œ ì—†ìŒ")
        except Exception as e:
            logger.error(f"API í˜¸ì¶œ ì‹¤íŒ¨: {e}")
            raise
        
        # ë‹¤ìŒ ë‹¬ë¡œ ì´ë™
        current_start = current_end + timedelta(seconds=1)
    
    return merged_results

def extract_file_url(self, item: Dict) -> Optional[str]:
        """
        API ì‘ë‹µ í•­ëª©ì—ì„œ ì²¨ë¶€ íŒŒì¼ URL ì¶”ì¶œ.

        Args:
            item: API ì‘ë‹µ í•­ëª© ë”•ì…”ë„ˆë¦¬.

        Returns:
            str ë˜ëŠ” None: ì²¨ë¶€ íŒŒì¼ URL ë˜ëŠ” None.
        """
        self.logger.info(f"ì²¨ë¶€ íŒŒì¼ URL ì¶”ì¶œ ì‹œì‘: {item}")
        file_url = item.get('atchFileUrl')
        if file_url and isinstance(file_url, str) and file_url.strip():
            return file_url.strip()
        return None

def download_file(self, file_url: str, save_dir: Optional[str] = None) -> str:
    """
    íŒŒì¼ì„ ë‹¤ìš´ë¡œë“œí•˜ì—¬ ì§€ì •ëœ ë””ë ‰í† ë¦¬ì— ì €ì¥.

    Args:
        file_url: ë‹¤ìš´ë¡œë“œí•  íŒŒì¼ì˜ URL.
        save_dir: íŒŒì¼ì„ ì €ì¥í•  ë””ë ‰í† ë¦¬ ê²½ë¡œ.
                  ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ì˜ "temp" í´ë”ì— ì €ì¥.

    Returns:
        str: ì €ì¥ëœ íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œ.

    Raises:
        requests.HTTPError: ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨ ì‹œ.
    """
    self.logger.info(f"íŒŒì¼ ë‹¤ìš´ë¡œë“œ ì‹œì‘: {file_url}")
    response = requests.get(file_url, stream=True)
    response.raise_for_status()

    # ë””ë ‰í† ë¦¬ ìƒì„±, ì—†ìœ¼ë©´ í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ì—ì„œ temp í´ë” ìƒì„±
    if save_dir:
        Path(save_dir).mkdir(parents=True, exist_ok=True)
    else:
        save_dir = os.path.join(os.getcwd(), "temp")
        Path(save_dir).mkdir(parents=True, exist_ok=True)
        # íŒŒì¼ ì´ë¦„ ì¶”ì¶œ
        file_name = os.path.basename(file_url)
        file_path = os.path.join(save_dir, file_name)

        # íŒŒì¼ ì €ì¥
        with open(file_path, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)

        self.logger.info(f"íŒŒì¼ ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {file_path}")
        return file_path
```

```python
import os
import requests
import re
import urllib.parse
from pathlib import Path
from typing import Optional

def download_file(file_urls: list[str], save_dir: Optional[str] = None) -> str:
    """
    íŒŒì¼ì„ ë‹¤ìš´ë¡œë“œí•˜ì—¬ ì§€ì •ëœ ë””ë ‰í† ë¦¬ì— ì €ì¥.

    Args:
        file_urls: ë‹¤ìš´ë¡œë“œí•  íŒŒì¼ë“¤ì˜ URL ë¦¬ìŠ¤íŠ¸.
        save_dir: íŒŒì¼ì„ ì €ì¥í•  ë””ë ‰í† ë¦¬ ê²½ë¡œ.
                  ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ì˜ "temp" í´ë”ì— ì €ì¥.
    Returns:
        str: ë§ˆì§€ë§‰ìœ¼ë¡œ ì €ì¥ëœ íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œ.

    Raises:
        requests.HTTPError: ë‹¤ìš´ë¡œë“œ ì‹¤íŒ¨ ì‹œ.
    """
    
    # ë””ë ‰í† ë¦¬ ì„¤ì •: ì§€ì •í•˜ì§€ ì•Šìœ¼ë©´ í˜„ì¬ ì‘ì—… ë””ë ‰í† ë¦¬ì˜ "temp" í´ë”ì— ì €ì¥
    if not save_dir:
        save_dir = Path(os.getcwd()) / "temp"
    else:
        save_dir = Path(save_dir)
        
    # ë””ë ‰í† ë¦¬ ìƒì„±
    save_dir.mkdir(parents=True, exist_ok=True)
    
    final_file_path = "" # ë§ˆì§€ë§‰ìœ¼ë¡œ ì €ì¥ëœ íŒŒì¼ì˜ ê²½ë¡œë¥¼ ë°˜í™˜í•˜ê¸° ìœ„í•¨
    
    for file_url in file_urls:
        print(f"íŒŒì¼ ë‹¤ìš´ë¡œë“œ ì‹œì‘: {file_url}")
        
        # 1. íŒŒì¼ ë‹¤ìš´ë¡œë“œ ìš”ì²­
        response = requests.get(file_url, stream=True)
        response.raise_for_status() # HTTP ì˜¤ë¥˜ê°€ ë°œìƒí•˜ë©´ ì˜ˆì™¸ ë°œìƒ

        # 2. íŒŒì¼ ì´ë¦„ ì¶”ì¶œ ë¡œì§ ìˆ˜ì •
        
        # a) Content-Disposition í—¤ë”ì—ì„œ íŒŒì¼ ì´ë¦„ ì¶”ì¶œ ì‹œë„
        if 'Content-Disposition' in response.headers:
            # 'filename=' ë˜ëŠ” 'filename*=' íŒ¨í„´ìœ¼ë¡œ íŒŒì¼ ì´ë¦„ì„ ì¶”ì¶œí•©ë‹ˆë‹¤.
            fname = re.findall(r'filename=["\']?([^"\']*)["\']?', response.headers['Content-Disposition'])
            # URL ë””ì½”ë”© ë° ì„¸ë¯¸ì½œë¡  ì œê±° ë¡œì§ ì¶”ê°€
            file_name_raw = urllib.parse.unquote(fname[0]) if fname else os.path.basename(file_url).split('?')[0]
        else:
            # b) í—¤ë”ê°€ ì—†ìœ¼ë©´ URL ê²½ë¡œì˜ ê¸°ë³¸ ì´ë¦„ ì‚¬ìš© ë° ì¿¼ë¦¬ ë¬¸ìì—´ ì œê±°
            file_name_raw = os.path.basename(file_url).split('?')[0] # '?' ì´í›„ì˜ ì¿¼ë¦¬ ë¬¸ìì—´ ì œê±°

        # ì¶”ê°€: íŒŒì¼ ì´ë¦„ ì •ë¦¬ (ìœ íš¨í•˜ì§€ ì•Šì€ ë¬¸ìë‚˜ ëì˜ ì„¸ë¯¸ì½œë¡  ì œê±°)
        file_name = file_name_raw.replace(';', '')
        
        # 3. ìµœì¢… ì €ì¥ ê²½ë¡œ ì„¤ì •
        # final_file_pathëŠ” Path ê°ì²´ì…ë‹ˆë‹¤.
        final_file_path = save_dir / file_name

        # 4. ê²½ë¡œ ê¸¸ì´ ì²´í¬ ë° Long Path Prefix ì ìš© (Windows)
        file_path_str = str(final_file_path)
        final_file_path_long = file_path_str # ê¸°ë³¸ê°’ ì„¤ì •

        # Windows í™˜ê²½ì—ì„œ ê²½ë¡œê°€ 260ì ì´ìƒì¼ ê²½ìš° ì ‘ë‘ì‚¬ë¥¼ ë¶™ì—¬ì¤ë‹ˆë‹¤.
        if len(file_path_str) > 260 and os.name == 'nt':
            # \\?\C:\... í˜•íƒœì˜ Long Path Prefix ì¶”ê°€
            if not file_path_str.startswith('\\\\?\\'):
                # ì ˆëŒ€ ê²½ë¡œë¡œ ë³€í™˜í•˜ì—¬ Long Path Prefixë¥¼ ì¶”ê°€í•©ë‹ˆë‹¤.
                final_file_path_long = '\\\\?\\' + os.path.abspath(file_path_str)
            # else: ì´ë¯¸ \\\\?\\ë¡œ ì‹œì‘í•˜ëŠ” ê²½ìš° final_file_path_longì€ file_path_strì„ ìœ ì§€
        
        # 5. íŒŒì¼ ì €ì¥ (***ë°˜ë“œì‹œ for ë£¨í”„ ì•ˆì— ìˆì–´ì•¼ í•©ë‹ˆë‹¤.***)
        with open(final_file_path_long, 'wb') as f:
            for chunk in response.iter_content(chunk_size=8192):
                f.write(chunk)
        
        print(f"íŒŒì¼ ë‹¤ìš´ë¡œë“œ ì™„ë£Œ: {final_file_path}")

    # ë§ˆì§€ë§‰ìœ¼ë¡œ ì €ì¥ëœ íŒŒì¼ì˜ ì „ì²´ ê²½ë¡œë¥¼ ë¬¸ìì—´ë¡œ ë°˜í™˜
    return str(final_file_path)

# ì‚¬ìš© ì˜ˆì‹œ
# file_urls_list = ["http://example.com/file1.pdf", "http://example.
```

```python
from datetime import datetime, timedelta
from typing import List, Dict
import re

# https://apis.data.go.kr/1230000/ad/BidPublicInfoService/getBidPblancListInfoCnstwk?serviceKey=8fea481d8c22a53e6baa58ac84350a16577b68d4ba8ee4b2c5ce3057db4bbbfa&pageNo=1&numOfRows=10&inqryDiv=1&inqryBgnDt=202511150000&type=json&inqryEndDt=202511170000&bidNtceNo=20200100008
def getBidPblancListInfoCnstwk(
    start_date: str, 
    end_date: str, 
    bid_ntce_no: str = '', 
    page_no: int = 1, 
    num_of_rows: int = 10, 
    inqry_div: int = 1
) -> List[Dict]:
    """
    ê³µì‚¬ ì…ì°° ê³µê³  ì •ë³´ë¥¼ ì¡°íšŒí•˜ë©°, 1ê°œì›” ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ì–´ í˜¸ì¶œ.

    Args:
        start_date: ì¡°íšŒ ì‹œì‘ ë‚ ì§œ (YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM).
        end_date: ì¡°íšŒ ì¢…ë£Œ ë‚ ì§œ (YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM).
        bid_ntce_no: ì…ì°° ê³µê³  ë²ˆí˜¸.
        page_no: í˜ì´ì§€ ë²ˆí˜¸.
        num_of_rows: í˜ì´ì§€ë‹¹ ê²°ê³¼ ìˆ˜.
        inqry_div: ì¡°íšŒ êµ¬ë¶„.

    Returns:
        List[Dict]: ë³‘í•©ëœ API ì‘ë‹µ ë°ì´í„° ë¦¬ìŠ¤íŠ¸.
        
    Raises:
        ValueError: ë‚ ì§œ í˜•ì‹ ì˜¤ë¥˜.
    """
    # ë‚ ì§œ í˜•ì‹ ê²€ì¦
    date_pattern = re.compile(r'^\d{8}(\d{4})?$')
    if not date_pattern.match(start_date):
        logger.error(f"start_date í˜•ì‹ ì˜¤ë¥˜: {start_date}")
        raise ValueError(f"start_dateëŠ” YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM í˜•ì‹ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    if not date_pattern.match(end_date):
        logger.error(f"end_date í˜•ì‹ ì˜¤ë¥˜: {end_date}")
        raise ValueError(f"end_dateëŠ” YYYYMMDD ë˜ëŠ” YYYYMMDDHHMM í˜•ì‹ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    
    # YYYYMMDD â†’ YYYYMMDDHHMM ë³€í™˜
    if len(start_date) == 8:
        start_date += '0000'
    if len(end_date) == 8:
        end_date += '0000'
    
    # ë‚ ì§œ ë³€í™˜
    start_dt = datetime.strptime(start_date, "%Y%m%d%H%M")
    end_dt = datetime.strptime(end_date, "%Y%m%d%H%M")
    if start_dt > end_dt:
        raise ValueError("start_dateëŠ” end_dateë³´ë‹¤ ì´ì „ì´ì–´ì•¼ í•©ë‹ˆë‹¤")
    
    # ê²°ê³¼ ë³‘í•©
    merged_results = []
    current_start = start_dt

    while current_start < end_dt:
        # 1ê°œì›” ë‹¨ìœ„ë¡œ ë‚˜ëˆ„ê¸°
        current_end = (current_start + timedelta(days=31)).replace(day=1) - timedelta(seconds=1)
        if current_end > end_dt:
            current_end = end_dt
        
        # ë‚ ì§œë¥¼ ë¬¸ìì—´ë¡œ ë³€í™˜
        current_start_str = current_start.strftime("%Y%m%d%H%M")
        current_end_str = current_end.strftime("%Y%m%d%H%M")
        
        logger.debug(f"API í˜¸ì¶œ: {current_start_str} ~ {current_end_str}")
        
        # API í˜¸ì¶œ
        try:
            data = get_data_go_kr(
                api_url=(
                    f"https://apis.data.go.kr/1230000/ad/BidPublicInfoService/"
                    f"getBidPblancListInfoCnstwk"
                    f"?pageNo={page_no}"
                    f"&numOfRows={num_of_rows}"
                    f"&inqryDiv={inqry_div}"
                    f"&type=json"
                    f"&bidNtceNo={bid_ntce_no}"
                    f"&inqryBgnDt={current_start_str}"
                    f"&inqryEndDt={current_end_str}"
                ),
                timeout=10
            )
            items = data.get('response', {}).get('body', {}).get('items', [])
            if items:
                merged_results.extend(items)
                logger.info(f"{current_start_str} ~ {current_end_str}: {len(items)}ê±´ ì¡°íšŒ")
            else:
                logger.info(f"{current_start_str} ~ {current_end_str}: ì¡°íšŒëœ ë¬¸ì„œ ì—†ìŒ")
        except Exception as e:
            logger.error(f"API í˜¸ì¶œ ì‹¤íŒ¨: {e}")
            raise
        
        # ë‹¤ìŒ ë‹¬ë¡œ ì´ë™
        current_start = current_end + timedelta(seconds=1)
    
    return merged_results
```
